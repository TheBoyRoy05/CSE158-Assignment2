```python
# load data
def load_gzip_data(data, file_name):
    with gzip.open(f"{file_name}.json.gz", "rt", encoding="utf-8") as f:
        for line in f:
            data.append(json.loads(line))
```

```python
# Write filtered data to review-SanDiego_10.json.gz
def write_gzip_data(data, file_name):
    with gzip.open(f'{file_name}.json.gz', 'wt', encoding='utf-8') as f:
        for d in tqdm(data):
            f.write(json.dumps(d) + '\n')
```

```python
reviews_sd = list()
for i in range(1,5):
    load_gzip_data(reviews_sd, f'Data/review-SanDiego_10_file{i}')
```
