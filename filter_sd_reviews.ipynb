```python
import numpy as np
import gzip
import json
```

```python
meta = []
with gzip.open("meta-California.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
        meta.append(json.loads(line))

reviews = []
with gzip.open("review-California_10.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
        reviews.append(json.loads(line))
```

```python
meta_sd = []
with gzip.open("meta-SanDiego.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
        meta_sd.append(json.loads(line))
```

```python
businesses_sd = set([meta_sd[i]['gmap_id'] for i in range(len(meta_sd))])
```

```python
reviews_sd = [r for r in reviews if r['gmap_id'] in businesses_sd]
```

```python
# Write filtered data to review-SanDiego_10.json.gz
with gzip.open('review-SanDiego_10.json.gz', 'wt', encoding='utf-8') as f:
    for d in tqdm(reviews_sd):
        f.write(json.dumps(d) + '\n')
```
