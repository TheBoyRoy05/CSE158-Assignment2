```python
import gzip
import json
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
def parse(path):
  g = gzip.open(path, 'rt', encoding='utf-8')
  for l in g:
    if not l: continue
    yield json.loads(l)
```

```python
with open('sd_zips.txt', 'r') as f:
    sd_zips = f.read().splitlines()
```

```python
data = [d for d in tqdm(parse('Data/meta-California.json.gz')) 
        if d and 'address' in d and d['address'] and len(d['address']) >= 5 and d['address'][-5:] in sd_zips]
```

```python
len(data), data[0]
```

```python
# Write filtered data to meta-SanDiego.json.gz
with gzip.open('Data/meta-SanDiego.json.gz', 'wt', encoding='utf-8') as f:
    for d in tqdm(data):
        f.write(json.dumps(d) + '\n')
        
print(f"Written {len(data)} records to meta-SanDiego.json.gz")
```
